```python
import sys
import matplotlib.pyplot as plt
import numpy as np

sys.path.append('../')
from vd_env.run_e_greedy_policy import run_e_greedy_policy
from vd_env.lambda_env import get_simple_env
```

## Set params for testing

```python
count_terms = 1_000
max_term_reward = 30
exploration = 50  # 2 strategies, so 50*2=100 total terms for exploration

epsilon = .1
```

## Run determenistic policy

```python
env = get_simple_env(count_terms=count_terms, max_step_term=max_term_reward)
_, rewards_det = run_e_greedy_policy(env, exploration=15, max_term_reward=max_term_reward,
                                     deterministic=True)
env.render()
```

## Run non-determenistic policy

```python
env.reset_soft()
_, rewards_non_det = run_e_greedy_policy(env, exploration=15, max_term_reward=max_term_reward,
                                         deterministic=False)
env.render()
```

## Draw some comparison

```python
rewards_det = [max_term_reward - rw for rw in rewards_det]
rewards_non_det = [max_term_reward - rw for rw in rewards_non_det]

cum_rewards_det = np.cumsum(rewards_det)
for i in range(len(cum_rewards_det)):
    cum_rewards_det[i] /= i + 1

cum_rewards_non_det = np.cumsum(rewards_non_det)
for i in range(len(cum_rewards_non_det)):
    cum_rewards_non_det[i] /= i + 1
```

```python
plt.plot(range(1, len(cum_rewards_det) + 1), cum_rewards_det, label="Deterministic")
plt.plot(range(1, len(cum_rewards_non_det) + 1), cum_rewards_non_det, label="not_Deterministic")
plt.legend(loc='upper right')
plt.title('E Greedy Policy')
plt.xlabel('Trials')
plt.ylabel('Avg Steps')
plt.show()
```

```python
start_point = 100

plt.plot(range(1, len(cum_rewards_det[start_point:]) + 1),
         cum_rewards_det[start_point:], label="Deterministic")
plt.plot(range(1, len(cum_rewards_non_det[start_point:]) + 1),
         cum_rewards_non_det[start_point:], label="not_Deterministic")
plt.legend(loc='upper right')
plt.title('E Greedy Policy')
plt.xlabel('Trials')
plt.ylabel('Avg Steps')
plt.show()
```

```python
import random
import sys

sys.path.append('../')

from vd_env.run_e_greedy_policy import run_e_greedy_policy
from vd_env.lambda_env import get_simple_env
import matplotlib.pyplot as plt
import numpy as np
```

## Set params for testing

```python
# set random seed for experiments reproduction
seed = 42

count_terms = 1_000
max_term_reward = 30
exploration = 50  # 2 strategies, so 50*2=100 total terms for exploration

epsilon = .1
```

## Run testing

```python
env = get_simple_env(count_terms=count_terms, max_step_term=max_term_reward)
_, rewards = run_e_greedy_policy(env, exploration=exploration,
                                 max_term_reward=max_term_reward,
                                 epsilon=epsilon)
env.render()
```

## Plot results:

```python
plt.plot(rewards)
plt.title('E Greedy Policy')
plt.xlabel('Trials')
plt.ylabel('Reward')
plt.show()
```

```python
cum_rewards = np.cumsum(rewards)
plt.plot(cum_rewards)
plt.title('E Greedy Policy')
plt.xlabel('Trials')
plt.ylabel('Cum Reward')
plt.show()
```

```python
for i in range(len(cum_rewards)):
    cum_rewards[i] /= i + 1
plt.plot(cum_rewards)
plt.title('E Greedy Policy')
plt.xlabel('Trials')
plt.ylabel('Avg Reward')
plt.show()
```

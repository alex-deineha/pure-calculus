```python
import numpy as np
import pandas as pd

from tqdm import tqdm

from matplotlib import pyplot as plt
from matplotlib import rcParams
```

```python
data_mae = pd.read_csv("../model_steps_maes.csv")
data_mae.head()
```

```python
rcParams["figure.figsize"] = 4, 3


def draw_lines(curves: dict, title="CONV, LO", axis_name="MAE"):
    plt.plot(range(len(curves["train"])), curves["train"], label="train", linestyle="--")
    plt.plot(range(len(curves["test"])), curves["test"], label="test", linestyle="-")

    plt.xlabel("steps")
    plt.ylabel(axis_name)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
draw_lines({
    "train": data_mae["MAE LO conv train"],
    "test": data_mae["MAE LO conv test"]
}, title="CONV, LO")
```

```python
draw_lines({
    "train": data_mae["MAE LO LSTM train"],
    "test": data_mae["MAE LO LSTM test"]
}, title="LSTM, LO")
```

```python
draw_lines({
    "train": data_mae["MAE LO transformer train"],
    "test": data_mae["MAE LO transformer test"]
}, title="Transformer, LO")
```

```python
draw_lines({
    "train": data_mae["MAE RI conv train"],
    "test": data_mae["MAE RI conv test"]
}, title="CONV, RI")
```

```python
draw_lines({
    "train": data_mae["MAE RI LSTM train"],
    "test": data_mae["MAE RI LSTM test"]
}, title="LSTM, RI")
```

```python
draw_lines({
    "train": data_mae["MAE RI transformer train"],
    "test": data_mae["MAE RI transformer test"]
}, title="Transformer, RI")
```
